```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import LSTM, Dropout, Dense, Activation
from keras.callbacks import TensorBoard, ModelCheckpoint, ReduceLROnPlateau

import datetime
import os
```

## 데이터셋 로드

사전 생성한 데이터셋을 불러온다. 데이터셋은 CoinMarketCap(https://coinmarketcap.com)에서 Ethereum Historical data를 통하여 획득하였다.

pandas의 read_csv 함수를 사용하여 csv 파일을 불러온다. 이때 0번 column은 날짜로 index로 지정하여 나머지 값이 float 형태로 입력되도록 한다.

```python
data = pd.read_csv('dataset/ethereum_dataset_2.csv', index_col=0)
data.head()
```

## 입력값 정규화

학습 수행 전에 데이터 범위를 정규화(data normalize)한다. 이는 local optimize에 빠지는 것을 줄이고 학습을 빠르게 하는데 도움이 된다.

수식: (요소값 - 최소값) / 최대값 - 최소값)

결과 전체 구간을 0~1 사이로 맞출 수 있다.

### 데이터 전처리

price의 중간값 mid_prices를 계산하여 저장한다.

RNN의 데이터는 시퀀셜 데이터로 구간(seq_len)에 따라 새로운 입력값이 순차적으로 들어온 데이터 형태를 갖는다.

result 리스트를 만들고 정규화를 진행하기 위한 시퀀셜 데이터를 생성한다.

```python
high_prices = data['High'].values
low_prices = data['Low'].values

mid_prices = (high_prices + low_prices) / 2

seq_len = 30
sequence_length = seq_len + 1

result = []
for index in range(len(mid_prices) - sequence_length):
    result.append(mid_prices[index: index + sequence_length])
```

result의 결과를 확인하면 다음 array마다 0번째 요소가 지워지고 -1번째 요소로 새로운 데이터(mid_price)가 입력되는 것을 확인할 수 있다.

```python
result
```

### 정규화 함수

데이터를 받으면 정규화된 데이터(normalized_data)를 반환하는 함수를 생성한다.

```python
def normalize_windows(data):
    normalized_data = []
    for window in data:
        normalized_window = [((float(p) / float(window[0])) - 1) for p in window]
        normalized_data.append(normalized_window)
    return np.array(normalized_data)
```

전처리 데이터를 정규화 함수에 입력하여 정규화 된 결과값을 얻는다.

```python
norm_result = normalize_windows(result)
```

정규화를 진행한 result의 일부 

```python
print(norm_result[0])
print(norm_result[1])
print(norm_result[3])
print(norm_result[4])
print(norm_result[5])
```

## 학습 데이터 및 테스트 데이터 분할

전체 학습 데이터에서 앞의 95%를 학습, 마지막 5%를 테스트 데이터로 나눈다.

train 데이터는 train에 정규화된 입력값을 넣고 shuffle 함수로 순서를 무작위로 섞어 학습을 진행한다.

test 데이터는 train 데이터로 사용하고 남은 나머지 데이터를 사용하며 이때는 순서를 섞지 않는다.

```python
train_cut = int(round(norm_result.shape[0] * 0.9))

train = norm_result[:train_cut, :]
np.random.shuffle(train)

x_train = train[:, :-1]
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
y_train = train[:, -1]

x_test = norm_result[train_cut:, :-1]
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
y_test = norm_result[train_cut:, -1]

x_train.shape, x_test.shape
```

## 모델 체크포인터 롤백
ModelCheckpoint 콜백 함수는 Keras 에서 모델을 학습할 때마다 중간중간 콜백 형태로 알려준다.

ModelCheckpoint의 속성으로 verbose는 해당 함수의 진행 사항의 출력여부, save_best_only는 모델의 정확도가 최고값을 갱신했을때만 저장하는 옵션이다.

from keras.callbacks import ModelCheckpoint

```python
MODEL_SAVE_FOLBER_PATH = './model/'
if not os.path.exists(MODEL_SAVE_FOLBER_PATH):
    os.mkdir(MODEL_SAVE_FOLBER_PATH)
    
model_path = MODEL_SAVE_FOLBER_PATH + '{epoch:02d}-{val_loss:.4f}.hdf5'

# 체크포인트 생성 코드
cb_checkpoint = ModelCheckpoint(filepath=model_path, monitor='val_loss',verbose=1, save_best_only=True, mode='auto')
```

## 모델 설계
   
케라스 Sequential()클래스를 model에 주어 딥러닝 아키텍처 설계를 시작한다.

케라스는 순서대로 모델을 쌓아 올리면서 아키텍처를 쉽게 구현할 수 있다.

    from keras.layers import LSTM, Dropout, Dense, Activation

케라스의 LSTM 레이어를 불러오면 쉽게 RNN 설계를 진행할 수 있다.

    keras.layers.LSTM(units, activation='tanh', recurrent_activation='hard_sigmoid', use_bias=True, kernel_initializer='glorot_uniform', recurrent_initializer='orthogonal', bias_initializer='zeros', unit_forget_bias=True, kernel_regularizer=None, recurrent_regularizer=None, bias_regularizer=None, activity_regularizer=None, kernel_constraint=None, recurrent_constraint=None, bias_constraint=None, dropout=0.0, recurrent_dropout=0.0, implementation=1, return_sequences=False, return_state=False, go_backwards=False, stateful=False, unroll=False)
    
    units: Positive integer. 출력 space의 차원을 정한다.
    return_sequences: Boolean. 출력 시퀀스의 마지막 출력을 반환할지 전체 시퀀스를 반환할지를 지정한다.
    input_shape: 처음 데이터를 입력받는 shape를 지정한다.
    
    참고: https://keras.io/layers/recurrent/

우리가 생성한 학습 시퀀셜 데이터는 (1305, 50, 1)의 구조를 가지며 첫 LSTM 의 input_shape를 이에 맞춰 (50, 1)로 지정한다.

케라스의 Dense 레이어는 입출력을 모두 연결해주는 레이어로 Fully connected 레이어라고 생각하면 이해하기 편하다.

    keras.layers.Dense(units, activation=None, use_bias=True, kernel_initializer='glorot_uniform', bias_initializer='zeros', kernel_regularizer=None, bias_regularizer=None, activity_regularizer=None, kernel_constraint=None, bias_constraint=None)
    
    units: Positive integer, 출력 space의 차원
    input_dim: 입력 차원 설정.
    init: 가중치 초기화 방법을 설정
        uniform: 균일 분포
        normal: 가우시안 분포
    activation: 활성화 함수로 무엇을 사용할지 지정한다. 아무것도 안적으면 아무것도 적용이 안됨. 
        linear: 기본값, 입력 뉴런과 가중치로 계산된 결과값이 그대로 출력으로 나옴.
        relu: rectifier 함수로 은닉층에 주로 사용.
        sigmoid: 이진 분류 문제 출력층에 주로 사용.
        softmax: 다중 클래스 분류 문제에서 출력층에 주로 사용.
        
여기서는 결과 출력이 1개이므로 unit을 1로 하였다.

```python
model = Sequential()

for i in range (4):
    model.add(LSTM(30, return_sequences=True, input_shape=(30, 1)))   

model.add(LSTM(30, return_sequences=False))

model.add(Dense(1, activation='linear'))

model.compile(loss='mse', optimizer='rmsprop')

model.summary()
```

```python
start_time = datetime.datetime.now().strftime('%Y_%m_%d_%H_%M_%S')

model.fit(x_train, y_train, validation_data=(x_test, y_test), batch_size=10, epochs=20, callbacks=[cb_checkpoint])
```

```python
from keras.models import load_model

# 모델 저장
model. save('ether_prediction_model.h5')

# 모델 불러오기
# model = load_model('ether_prediction_model.h5')
```
